# Algorithme de calcul de distance

```python
import numpy as np
def haversine_vectorize(lon1, lat1, lon2, lat2):
 
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
 
    newlon = lon2 - lon1
    newlat = lat2 - lat1
 
    haver_formula = np.sin(newlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(newlon/2.0)**2
 
    dist = 2 * np.arcsin(np.sqrt(haver_formula ))
    km = 6367 * dist #6367 for distance in KM for miles use 3958
    return km
```

## Calcule des distances entre stations et les gares 

```python
liste_gare = list(df_distance['nom_gare'])
liste_station = list(df_distance2['stationcode'])

df = pd.DataFrame()
df['stationcode'] = liste_station

for i in liste_gare:
    lon2 = df_distance.loc[df_distance['nom_gare']== i, 'longitude'].values[0]
    lat2 = df_distance.loc[df_distance['nom_gare']== i, 'latitude'].values[0]
    distance = []
    for j in liste_station:
        lon1 = df_distance2.loc[df_distance2['stationcode']==j, 'longitude'].values[0]
        lat1 = df_distance2.loc[df_distance2['stationcode']==j, 'latitude'].values[0]
        distance.append(haversine_vectorize(lon1, lat1, lon2, lat2))
        #print(distance)
    df[i] = distance
```
