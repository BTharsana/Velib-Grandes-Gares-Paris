# Importation des libraires 

```python
import pandas as pd
```

# Chargement du fichier source

```python
df_avril = pd.read_csv('Velib_Avril_2022.csv')
```

```python
df_avril.shape
```

```python
df_avril.info()
```

```python
df_avril.head()
```

```python
df_avril.isnull().sum()
```

```python
df_avril.loc[df_avril['name'].isnull()]
```

Nous avons remarqué que ces stations n'ont pas encore était installé et qu'elles ne nous apportent pas plus d'informations. De ce fait, nous avons décidé de les supprimer.

```python
df_avril = df_avril.dropna()
```

```python
df_avril.isnull().sum()
```

```python
df_avril['ping_time'].str[:10].unique()
```

On remarque que nous avons des problèmes de récupération de données. Pour des collectes qui ont été faite en mai ou en juin des échantillons d'avril ont été fourni. Nous faisons le choix de supprimer ces lignes suplémentaire qui apporte un biais à notre dataset.

```python
#ping time différents d'avril
index_hors_avril = df_avril.loc[df_avril['ping_time'].str[3:-14] != '04'].index
df_avril=df_avril.drop(index_hors_avril)
```

```python
df_avril.shape
```

## Remarque

```python
df_avril.groupby('ping_time')['stationcode'].count().sort_values()
```

Il y a 1419 stations uniques mais elles ne sont pas toutes enregistrés pour les différents ping time.

## Doublons

```python
df_avril.duplicated().sum()
```

```python
df_avril.loc[df_avril.drop(columns='ping_time').duplicated()]
```

Nous avons 2204974 lignes doublons dans notre dataset que nous avons choisi de supprimés

```python
#On récupère les index des lignes qui sont doublons et  nous les supprimons par la suite
dupli_index = df_avril.loc[df_avril.drop(columns='ping_time').duplicated()].index
df_avril = df_avril.drop(dupli_index)
```

```python
df_avril.loc[df_avril.drop(columns='ping_time').duplicated()]
```

```python
df_avril.shape
```

### Suppression Romainville - Vaillant-couturier

```python
df_avril.loc[df_avril['coordonnees_geo']=='[48.868054, 2.4470495]']
```

Comme la capacité est égale à 0, nous avons choisi de supprimer les données

```python
romainville_index = df_avril.loc[df_avril['name']=='Romainville - Vaillant-couturier'].index
df_avril = df_avril.drop(romainville_index)
```

### Correction des erreurs d'orthographe dans les noms de station

```python
dict_nom = {'Edouard Tremblay - Julian Grimau' : 'Edouard Tremblay - Julian Grimaux',
            'Stade Georges Carpentier' : 'Station Paris Basketball - Stade Georges Carpentier'}

df_avril['name'] = df_avril['name'].replace(dict_nom)
```

```python
df_avril[['name','stationcode']].loc[df_avril['stationcode']=='44013']
```

## Format date

```python
df_avril['duedate'] = pd.to_datetime(df_avril['duedate'])
df_avril['duedate'] = df_avril['duedate'].dt.strftime('%Y-%m-%d %H:%M:%S')
```

# Création de colonne

```python
#Ratio des disponibilités : ([Nb bornes disponibles] / [Nombres de bornes en station])
df_avril['dispo_velo'] = df_avril['numbikesavailable']/df_avril['capacity']
df_avril['seuil_velo'] = 1/df_avril['capacity']
df_avril['dispo_bornette'] = df_avril['numdocksavailable']/df_avril['capacity']

#capcité des stations plus que les valeurs théorique attendus 
df_avril['capacity_calc'] = df_avril['numbikesavailable'] + df_avril['numdocksavailable']
```

```python
df_avril.shape
```

```python
df_avril.loc[df_avril['capacity']!=(df_avril['numbikesavailable'] + df_avril['numdocksavailable'])]
```

# Exportation Finale

```python
df_avril.shape
```

```python
df_avril.to_csv('Avril_2022_clean.csv', index=False)
print('***Exported***')
```

## Répartition des stations pour faciliter la préparation

```python
list_station = list(df_avril['stationcode'].unique())
list_station[:472]
```

```python
#Répartition Kamala
df_kamala = df_avril.loc[df_avril['stationcode'].apply(lambda x: x in list_station[:472])]

df_kamala.to_csv('df_kamala.csv', index=False)
```

```python
df_nassriat = df_avril.loc[df_avril['stationcode'].apply(lambda x: x in list_station[472:945])]
df_nassriat.to_csv('df_nassriat.csv', index=False)
```

```python
df_tharsana = df_avril.loc[df_avril['stationcode'].apply(lambda x: x in list_station[945:])]
df_tharsana.to_csv('df_tharsana.csv', index=False)
```
